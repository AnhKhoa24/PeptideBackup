```python
import subprocess

input_fasta = "/content/output_fragments_284.fasta"
output_fasta = "output_fragments_284_clean.fasta"
threshold = 0.99
memory_limit = 4000  # MB

cmd = f"cd-hit -i {input_fasta} -o {output_fasta} -c {threshold} -M {memory_limit}"

try:
    result = subprocess.run(cmd, shell=True, check=True, capture_output=True, text=True)
    print(result.stdout)
    print(f"CD-HIT đã chạy xong! Kết quả lưu ở {output_fasta}")
except subprocess.CalledProcessError as e:
    print("Lỗi khi chạy CD-HIT:")
    print(e.stderr)
```
