```python
import matplotlib.pyplot as plt
import numpy as np

def fingerprint(peptide):
    # Các từ điển thuộc tính amino acid
    properties = {
        "hydrophobicity": {
            "polar": "RKEDQN",
            "neutral": "GASTPHY",
            "hydrophobic": "CLVIMFW"
        },
        "vdw_volume": {
            "small": "GASTPDC",
            "medium": "NVEQIL",
            "large": "MHKFRYW"
        },
        "polarity": {
            "polar": "EDQNKR",
            "neutral": "GASTPHY",
            "nonpolar": "CLVIMFW"
        },
        "polarizability": {
            "low": "GASDT",
            "medium": "CPNVEQIL",
            "high": "KMHFRYW"
        },
        "charge": {
            "negative": "DE",
            "neutral": "ACFGHILMNPQSTVWY",
            "positive": "KR"
        },
        "secondary_structure": {
            "helix": "EALMQKRH",
            "sheet": "VIYCWT",
            "coil": "GNPSD"
        },
        "solvent_accessibility": {
            "buried": "ALFCGIVW",
            "intermediate": "MPSTHY",
            "exposed": "DEKNQR"
        }
    }

    # Hàm mã hóa màu sắc cho từng thuộc tính
    def get_color_for_property(aa, property_dict):
        for category, amino_acids in property_dict.items():
            if aa in amino_acids:
                if category == "polar" or category == "negative":
                    return "yellow"  # Polar and negative amino acids in red
                elif category == "neutral":
                    return "orange"  # Neutral amino acids in yellow
                elif category == "hydrophobic" or category == "nonpolar":
                    return "red"  # Hydrophobic and nonpolar amino acids in green
                elif category == "small":
                    return "yellow"  # Small volume amino acids in blue
                elif category == "medium":
                    return "orange"  # Medium volume in orange
                elif category == "large":
                    return "red"  # Large volume amino acids in purple
                elif category == "low":
                    return "yellow"  # Low polarizability in lightblue
                elif category == "medium":
                    return "orange"  # Medium polarizability in cyan
                elif category == "high":
                    return "red"  # High polarizability in dark blue
                elif category == "helix":
                    return "yellow"  # Helix in light green
                elif category == "sheet":
                    return "orange"  # Sheet in light coral
                elif category == "coil":
                    return "red"  # Coil in light yellow
                elif category == "buried":
                    return "yellow"  # Buried in gray
                elif category == "intermediate":
                    return "orange"  # Intermediate in pink
                elif category == "exposed":
                    return "red"  # Exposed in brown
        return "blue"  # Default for unmatched cases

    # Tạo các biểu đồ cho từng thuộc tính
    attributes = list(properties.keys())

    # Vẽ từng thuộc tính trên cửa sổ riêng biệt
    for i, attribute in enumerate(attributes):
        colors = [get_color_for_property(aa, properties[attribute]) for aa in peptide]
        
        plt.figure(figsize=(20, 1.5))
        plt.bar(range(len(peptide)), np.ones(len(peptide)), color=colors, tick_label=list(peptide))
        plt.title(attribute.capitalize())
        #plt.ylabel(attribute.capitalize())
        plt.tight_layout()
        plt.show()
```

```python

# Sử dụng hàm fingerprint với chuỗi peptide
peptide = "MKKKKYLIDLFSGCGGLSFGFEQAGFECLIGVDIEQSAL"
fingerprint(peptide)
```
